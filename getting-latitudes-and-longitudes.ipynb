# Getting Postal Codes related Coordinates

```python
import pandas as pd
df = pd.read_excel('cleaned_dataset.xlsx')
df.head()
```

I tried here to get the coordinates using Geocoder Library but it didn't complete all the runs and it timed out for hours with no progress so I got the coordinates using the file provided in the instructions 

```python
import geocoder # import geocoder

# initialize your variable to None
lats=[]
longs=[]
for i,row in df.iterrows():
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(row['Postcode']))
        lat_lng_coords = g.latlng
    lats.append(lat_lng_coords[0])
    longs.append(lat_lng_coords[1])
    print('{} from {} finished'.format(i,len(df)),end='\r')
```

```python
coords = pd.read_csv('Geospatial_Coordinates.csv')
coords.head()
```

```python
lats=[]
longs=[]
for i,row in df.iterrows():
    coord_row = coords.loc[coords['Postal Code']==row['Postcode']]
    lats.append(coord_row.iloc[0,1])
    longs.append(coord_row.iloc[0,2])
    print('{} from {} finished'.format(i,len(df)),end='\r')
```

```python
df['Latitude'] = lats
df['Longitude'] = longs
```

```python
df.sort_values('Neighborhood').reset_index(drop=True).drop('Unnamed: 0',axis=1).head(10)
```

```python
df.to_excel('canada_geo_dataset.xlsx')
```
